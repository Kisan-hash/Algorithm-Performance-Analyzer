```python
# ============================================
# 📊 Algorithm Performance Analyzer
# Comparing Sorting Algorithms
# ============================================

import time
import random
import matplotlib.pyplot as plt

print("Starting Algorithm Performance Analyzer 🚀")

# -------- 1️⃣ Sorting Algorithms --------

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


# -------- 2️⃣ Performance Testing --------

input_sizes = [100, 300, 500, 700, 1000]

bubble_times = []
merge_times = []
quick_times = []

for size in input_sizes:
    data = random.sample(range(10000), size)

    # Bubble Sort
    arr = data.copy()
    start = time.time()
    bubble_sort(arr)
    bubble_times.append(time.time() - start)

    # Merge Sort
    arr = data.copy()
    start = time.time()
    merge_sort(arr)
    merge_times.append(time.time() - start)

    # Quick Sort
    arr = data.copy()
    start = time.time()
    quick_sort(arr)
    quick_times.append(time.time() - start)

print("Performance Testing Completed ✅")


# -------- 3️⃣ Plot Results --------

plt.figure()
plt.plot(input_sizes, bubble_times)
plt.plot(input_sizes, merge_times)
plt.plot(input_sizes, quick_times)

plt.xlabel("Input Size")
plt.ylabel("Execution Time (seconds)")
plt.title("Sorting Algorithm Performance Comparison")
plt.legend(["Bubble Sort", "Merge Sort", "Quick Sort"])

plt.show()

print("Project Executed Successfully 🚀")
```
